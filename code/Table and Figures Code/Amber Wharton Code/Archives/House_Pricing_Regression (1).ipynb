```python
###Only Wharton
l = ['Single Unit Permits','Multi-Unit Permits','All Unit Permits','% Urban','Median Household Income_2021','Median Gross Rent_Percent_Change','Median Home Value (Owner Occupied)_Percent_Change','Median Gross Rent_2021','Median Home Value (Owner Occupied)_2021']
Wharton_R = []
Wharton_P = []
for i in l:

  import statsmodels.api as sm
  import pandas as pd

  data = merged_Wharton_price.dropna()##900+data
  # æ·»åŠ å¸¸æ•°åˆ—ä½œä¸ºå›žå½’æ¨¡åž‹çš„æˆªè·
  data = sm.add_constant(data)

  x = data[["WRLURI18"]]
  y = data[i]
  x = sm.add_constant(x)
  # åˆ›å»ºçº¿æ€§å›žå½’æ¨¡åž‹
  model = sm.OLS(y,x)

  # æ‹Ÿåˆæ¨¡åž‹
  results = model.fit()
  print(results.summary())

  # èŽ·å–p-valueå’Œr-squareå€¼
  p_value = results.pvalues[["WRLURI18"]]
  r_square = results.rsquared

  print("p-value:", p_value)
  Wharton_P.append(p_value)
  print("r-square:", r_square)
  Wharton_R.append(r_square)
```
