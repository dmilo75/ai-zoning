```python
import pandas as pd
r = pd.read_excel("round(2).xlsx")
r
```

```python
a = r.columns.tolist()
a
```

```python
len(r)
```

```python
for i in a:
  print(i)
  value = []
  change = r[i]
  for j in change:
    if j == "/":
      value.append("/")
    else:
      value.append(round(j,2))
  print(len(value))
  r[i] = value
r
```
